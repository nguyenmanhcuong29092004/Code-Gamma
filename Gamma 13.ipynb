```python
import nltk
from nltk import pos_tag, word_tokenize
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

def preprocess_text(text):
    # Tokenize văn bản
    words = word_tokenize(text)

    # Lọc stop words
    stop_words = set(stopwords.words('english'))
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words]

    return filtered_words
```

```python

# Tạo danh sách DataFrame
list_of_dataframes = [new41_data, new31_data, new21_data, new11_data]

def calculate_ratios(list_of_dataframes, words):
    ratios = {}
    filtered_rows_1_number = {word: 0 for word in words}
    filtered_rows_0_number = {word: 0 for word in words}
    
    for data in list_of_dataframes:
        for word in words:
            filtered_rows_1 = data[data['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x) & (data['label'] == 1)]
            filtered_rows_0 = data[data['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x) & (data['label'] == 0)]
            filtered_rows_1_number[word] += filtered_rows_1.shape[0]
            filtered_rows_0_number[word] += filtered_rows_0.shape[0]
        
    for word in words:
        ratio_1 = filtered_rows_1_number[word] / 40000
        ratio_0 = filtered_rows_0_number[word] / 40000

        ratios[word] = {'Ratio_1': ratio_1, 'Ratio_0': ratio_0}

    return ratios


# Lấy danh sách các từ từ DataFrame 'word_frequencies'
selected_words = filtered_df['Word'].tolist()

# Gọi hàm để tính tỷ lệ
ratios = calculate_ratios(list_of_dataframes, selected_words)

ratios_df = pd.DataFrame.from_dict(ratios, orient='index').reset_index()
ratios_df.columns = ['Word', 'Ratio_1', 'Ratio_0']
# Hợp hai DataFrame dựa trên cột 'Word'
result_df = pd.merge(filtered_df, ratios_df, on='Word')
result_df
```

```python
import pandas as pd

new111_data = new11_data.copy()
new211_data = new21_data.copy()
new311_data = new31_data.copy()
new411_data = new41_data.copy()

# Hàm tính xác suất cho từng từ trong danh sách
def calculate_probability_0(word):
    # Lọc các hàng có từ xuất hiện trong AdjectivesAndVerbsAndAdverb và label = 0
    word_rows_label_0 = pd.concat([
        new111_data[(new111_data['label'] == 0) & new111_data['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x)],
        new211_data[(new211_data['label'] == 0) & new211_data['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x)],
        new311_data[(new311_data['label'] == 0) & new311_data['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x)],
        new411_data[(new411_data['label'] == 0) & new411_data['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x)]
    ], ignore_index=True)
    
    # Tính tổng số hàng có từ trong AdjectivesAndVerbsAndAdverb
    total_count = sum(
        len(df[df['AdjectivesAndVerbsAndAdverb'].apply(lambda x: word in x)])
        for df in [new111_data, new211_data, new311_data, new411_data]
    )
    
    # Tính tỷ lệ
    probability_word_label_0 = len(word_rows_label_0) / total_count 

    return probability_word_label_0

# Tạo DataFrame mới
probability_data_0 = pd.DataFrame(columns=['Word', 'Probability_Label_0'])

# Thêm dữ liệu vào DataFrame
for word in combined_words:
    probability = calculate_probability_0(word)
    probability_data_0 = pd.concat([probability_data_0, pd.DataFrame({'Word': [word], 'Probability_Label_0': [probability]})], ignore_index=True)

# In kết quả
probability_data_0 = probability_data_0.sort_values(by='Probability_Label_0', ascending=False)
# Gộp hai DataFrame dựa trên cột 'Word'
probability_data_0 = pd.merge(filtered_test_adjectives_and_verbs_1, probability_data_0, on='Word', how='left')
# In kết quả
probability_data_0 = probability_data_0.sort_values(by='Probability_Label_0', ascending=False)
probability_data_0
```

```python
test2_data = test1_data.copy()


# Hàm tính xác suất cho mỗi văn bản
def calculate_text_probability_bayes(text, probabilities_df, prior_prob=0.5):
    words = text.split()
    log_odds_list = []

    for word in words:
        # Tìm xác suất của từ trong DataFrame probabilities_df
        word_prob_label_0 = probabilities_df[probabilities_df['Word'] == word]['Probability_Label_0'].values

        # Nếu từ không có trong DataFrame probabilities_df, gán xác suất mặc định
        if len(word_prob_label_0) == 0:
            word_prob_label_0 = prior_prob
        else:
            word_prob_label_0 = word_prob_label_0[0]

        # Tính toán log odds của từ
        log_odds = np.log(word_prob_label_0 / (1 - word_prob_label_0))
        log_odds_list.append(log_odds)

    # Kiểm tra nếu danh sách log odds khác 0 trước khi tính tổng
    if len(log_odds_list) > 0:
        # Tính tổng log odds của các từ trong văn bản
        sum_log_odds = sum(log_odds_list)
        
        # Tính xác suất cuối cùng cho văn bản dựa trên công thức Bayes
        text_probability = 1 / (1 + np.exp(-sum_log_odds))
    else:
        # Nếu danh sách log odds là 0, gán giá trị mặc định hoặc xử lý theo nhu cầu của bạn
        text_probability = prior_prob  # Giá trị mặc định
    
    return text_probability



test2_data['Text_Probability_Label_0'] = test2_data['AdjectivesAndVerbsAndAdverb'].apply(
    lambda x: calculate_text_probability_bayes(' '.join(x), probability_data_0) if isinstance(x, list) else 0.5
)
# Hiển thị DataFrame kết quả
```
